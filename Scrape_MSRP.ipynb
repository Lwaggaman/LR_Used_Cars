```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import time, random
```

```python
#Get links
makes = pd.read_csv('Data/make_model.csv')
links = ['https://www.cars.com/research/'+make+'/' for make in makes.Make]

#Get soups
soups = []
for link in links:
    res = requests.get(link)
    time.sleep(random.randint(0,4))
    if res.status_code == 200:
        soups.append(bs(res.text))
        print(links.index(link))
    else:
        print(f'error:{link}')
```

```python
#Function to get relevant data from HTML into a dict
def get_dict(soup):
    cars = [i.text for i in soup.find_all(class_="make-model-text")]
    msrps = [i.text for i in soup.find_all(class_='msrp-text--bold')]
    my_dict = dict(zip(cars, msrps))
    return my_dict

#Apply function to soups
data = dict()
for soup in soups:
    data.update(get_dict(soup))

#Save data
df = pd.DataFrame()
df['Model'] = data.keys()
df['msrp'] = data.values()
df.to_csv('Data/msrp_data.csv', ignore_ind)
```
